```python
def csv_to_json(file_path, json_file_path, columns_to_extract):
    """
    Convert a CSV file to a JSON file and extract specific columns.

    Parameters:
    - csv_file_path: str, path to the input CSV file
    - json_file_path: str, path to the output JSON file
    - columns_to_extract: list, list of column names to extract
    """
    data = []

    # Read the CSV file
    with open(file_path, mode='r', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        
        # Iterate over each row and extract the specified columns
        for row in csv_reader:
            extracted_data = {col: row[col] for col in columns_to_extract if col in row}
            data.append(extracted_data)

    # Write the data to a JSON file
    with open(json_file_path, mode='w', encoding='utf-8') as json_file:
        json.dump(data, json_file, indent=4)

file_path = file_path
json_file_path = r'C:\naibot\RAG_data\sentiment_data.json'
columns_to_extract = ['text', 'sentiment', "Platform"]  # Replace with your column names

csv_to_json(file_path, json_file_path, columns_to_extract)
```
